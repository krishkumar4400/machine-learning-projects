## Titanic Servival Prediction

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('titanic')
```

```python
df
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
df.size
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.drop(["deck", "embark_town", "alive", "class", "who", "adult_male"], axis=1, inplace=True)
```

```python
df.info()
```

```python
df["embarked"].value_counts()
```

```python
df['age'].fillna(df['age'].mean(), inplace=True)
```

```python
df.info()
```

```python
df.dropna(subset=["embarked"], inplace=True)
```

```python
df.info()
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder

encode = LabelEncoder()
```

```python
df['sex'] = encode.fit_transform(df['sex'])
df['embarked'] = encode.fit_transform(df['embarked']) # S=2, C=0, Q=1 
```

```python
df.head()
```

```python
df['alone'] = df['alone'].astype(int)
```

```python
df
```

```python
X = df.drop('survived', axis=1)
y = df['survived']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

### KNN

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
X_test_scaled
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors = 5)

knn_model.fit(X_train_scaled, y_train)
```

```python
knn_y_predict = knn_model.predict(X_test_scaled)
```

```python
accuracy_score(y_test, knn_y_predict)
```

```python
confusion_matrix(y_test, knn_y_predict)
```

```python
print(classification_report(y_test, knn_y_predict))
```

### Naive Bayes Algorithm

```python
from sklearn.naive_bayes import GaussianNB

model_naive_bayes = GaussianNB()
```

```python
model_naive_bayes.fit(X_train, y_train)
```

```python
y_pred_naive_bayes = model_naive_bayes.predict(X_test)
```

```python
y_pred_naive_bayes
```

```python
accuracy_score(y_test, y_pred_naive_bayes)
```

```python
confusion_matrix(y_test, y_pred_naive_bayes)
```

```python
print(classification_report(y_test, y_pred_naive_bayes))
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train_scaled, y_train)
```

```python
clf_predict = clf.predict(X_test_scaled)
```

```python
clf_predict
```

```python
accuracy_score(y_test, clf_predict)
```

```python
confusion_matrix(y_test, clf_predict)
```

```python
print(classification_report(y_test, clf_predict))
```

```python
# We use kernal to distinguise between red and green points

from sklearn.svm import SVC
model_svm = SVC(kernel="rbf")
model_svm.fit(X_train_scaled, y_train)
```

```python
svm_y_pred = model_svm.predict(X_test_scaled)
```

```python
svm_y_pred
```

```python
accuracy_score(y_test, svm_y_pred)
```

```python
confusion_matrix(y_test, svm_y_pred)
```

```python
print(classification_report(y_test, svm_y_pred))
```
